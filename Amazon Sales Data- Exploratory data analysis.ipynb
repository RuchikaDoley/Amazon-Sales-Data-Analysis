```python
#Importing all the necessary packages:
#Pandas- this package helps with reading the data from CSV file and is useful for data manipulation 
#with the help of dataframes.
import pandas as pd
#Numpy helps dealing with numerical calculations and array manipulation
import numpy as np
#Matplib helps to perform plotting and create bar charts, pie charts, scatterplots, etc.
import matplotlib.pyplot as plt
#Seaborn is a package with helps create clear, aesthetically pleasing and customisable visualizations
import seaborn as sns
#We import the date module from the datetime library so that we can handle the date columns better
from datetime import date
```

```python
#Loading dataset and looking at the first 10 rows of all columns
df=pd.read_csv('SALESDATA.csv')
df.head(10)
```

```python
#Dropping empty columns:
df=df.drop(['Unnamed: 20','Unnamed: 21','Unnamed: 22','Unnamed: 23'], axis=1)
df.head()
```

```python
#Getting more information about the datatypes and number of null values
df.info()
```

```python
#Checking the frequency of different item classes
df['Item Class'].value_counts()
```

It is clear that the vast majority of items fall under one class. A negligible number of items fall in other classes with majority of the classes having only 1 item sold. In addition there are a lot of missing values so it is evident that this column will not aid in analysis.

```python
#Checking if datekey and invoice date match
df[['DateKey','Invoice Date']]
```

It is clear that both of the columns match, making one of them redundant.

```python
#Dropping unwanted columns
df=df.drop(['Item Number','Line Number','Item Class','Invoice Date'], axis=1)
```

```python
df
```

```python
df['Invoice Number'].value_counts()
```

```python
#Dropping invoice number since it will not feature in further analysis.
df=df.drop("Invoice Number",axis=1)
```

```python
df['Sales Rep'].value_counts()
```

This shows the sales reps who were involved in the highest number of sales.

```python
df.head()
```

```python
#Converting DateKey and Promised Delivery Date to datetime in columns "Date" and "Delivery Date" to make it easy to 
#manipulate for analysis.
df['Date']=pd.to_datetime(df["DateKey"],dayfirst=True)
df["Delivery Date"]=pd.to_datetime(df["Promised Delivery Date"])
df=df.drop(["DateKey","Delivery Date"], axis=1)
df
```

```python
#Storing Year, month and Quarter in seperate columns
df["Year"]=df["Date"].dt.year
df["Month"]=df["Date"].dt.month
df["Quarter"]=df["Date"].dt.quarter
df
```

```python
#Checking the items which were purchased in the largest number of orders.
df['Item'].value_counts()
```

```python
#Checking the number of years of data and number of distinct purchases in each year.
df['Year'].value_counts()
```

```python
#Preparing series for each year containing the Months as Index and Total Sales Amount as values 
index_values=["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
sales_2017=df[df["Year"]==2017]
monthly_sales_2017=sales_2017.groupby('Month')['Sales Amount'].sum()
monthly_sales_2017.index=index_values[:sales_2017["Month"].value_counts().size]

sales_2018=df[df["Year"]==2018]
monthly_sales_2018=sales_2018.groupby('Month')['Sales Amount'].sum()
monthly_sales_2018.index=index_values[:sales_2018["Month"].value_counts().size]


sales_2019=df[df["Year"]==2019]
monthly_sales_2019=sales_2019.groupby('Month')['Sales Amount'].sum()
monthly_sales_2019.index=index_values[:sales_2019["Month"].value_counts().size]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Plotting data on each subplot
ax1.plot(monthly_sales_2017)
ax1.set_title('2017')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales')

ax2.plot(monthly_sales_2018)
ax2.set_title('2018')
ax2.set_xlabel('Month')

ax3.plot(monthly_sales_2019)
ax3.set_title('2019')
ax3.set_xlabel('Month')

# Displaying the plot
plt.tight_layout()
plt.show()
```

The graphs show that the total sales amount has been dropping over the years with a steep decline between March 2018 and January 2019. Towards the end of January the Sales make a recovery but still fails to reach the levels of 2017.

```python
#Preparing series for each year containing the Months as Index and Total Sales Margin Amount as values 
index_values=["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
salesmargin_2017=df[df["Year"]==2017]
monthly_salesmargin_2017=salesmargin_2017.groupby('Month')['Sales Margin Amount'].sum()
monthly_salesmargin_2017.index=index_values[:salesmargin_2017["Month"].value_counts().size]

salesmargin_2018=df[df["Year"]==2018]
monthly_salesmargin_2018=salesmargin_2018.groupby('Month')['Sales Margin Amount'].sum()
monthly_salesmargin_2018.index=index_values[:salesmargin_2018["Month"].value_counts().size]


salesmargin_2019=df[df["Year"]==2019]
monthly_salesmargin_2019=salesmargin_2019.groupby('Month')['Sales Margin Amount'].sum()
monthly_salesmargin_2019.index=index_values[:salesmargin_2019["Month"].value_counts().size]
```

```python
# Creating subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Plotting data on each subplot
ax1.bar(monthly_salesmargin_2017.index,monthly_salesmargin_2017.values, color='b', label='2017')
ax1.set_title('2017')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales Margin Amount')

ax2.bar(monthly_salesmargin_2018.index,monthly_salesmargin_2018.values, color='g', label='2018')
ax2.set_title('2018')
ax2.set_xlabel('Month')

ax3.bar(monthly_salesmargin_2019.index,monthly_salesmargin_2019.values, color='r', label='2019')
ax3.set_title('2019')
ax3.set_xlabel('Month')

# Displaying the plot
plt.tight_layout()
plt.show()
```

Evidently, Sales margin amount has followed a similar trend to Sales amount and shows a steep decline betwee March 2018 and January 2019. Overall the sales margin has fallen from the year 2017 and is making a steady recovery in the year 2019 although it is yet to make a full recovery back to the initial levels.

```python
marginsalesperitem=df.groupby("Item")["Sales Margin Amount"].sum().sort_values(ascending=False)
marginsalesperitem
```

By summing the total sales margin amount of each item, it is easy to decipher which items are making large profits and which ones are making losses. 
Certain items like Better Canned Shrimp and High Top Dried Mushrooms have massive sales margins($5459826.26 and $4659800.16 respectively). It helps that these items are among the top selling items across the years, thereby increasing profits.
On the other hand there are several loss making items like Fast Lemon Cookies and Best Choice Fondue Mix. Efforts will have to be made to develop more cost effective ways of production or to ease them out of the inventory.

```python
#Showing the contribution of the top 5 items in the total Sales Margin Amount.
top5=marginsalesperitem[:5]
other=pd.Series(marginsalesperitem[5:].sum())
itemvals=top5.append(other)
items=["Better Large Canned Shrimp","High Top Dried Mushrooms","Big Time Frozen Cheese Pizza","Fast Mini Donuts","Discover Manicotti","Other"]
plt.pie(x=itemvals.values,labels=items);
```

The top 5 items contribute almost 25% of the total sales margin amount collected from sales.

```python
correlation_matrix = df[["Discount Amount", "Item", "List Price", "Sales Margin Amount", "Sales Price", "Sales Quantity", "Sales Rep","Year","Month"]].corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,vmin=-1, vmax=1, cmap="YlGnBu")
plt.title('Correlation Heatmap')
plt.show()
```

Sales Margin Amount and Sales Quantity have a strong positive correlation.
Discount amount and Sales quantity do not have a strong positive correlation. This implies that discounts are not likely to have a major impact on sales(unless the discount amount is significantly high).
Sales Margin Amount and Discount Amount are positively correlated.
